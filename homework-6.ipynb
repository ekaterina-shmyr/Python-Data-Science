```python
#Тема “Обучение с учителем”

#Задание 1
#Импортируйте библиотеки pandas и numpy.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Загрузите "Boston House Prices dataset" из встроенных наборов данных библиотеки sklearn. 
from sklearn.datasets import load_boston
boston = load_boston()
boston.keys()
```

```python
#Создайте датафреймы X и Y из этих данных.
data = boston["data"]
data.shape
```

```python
feature_names = boston["feature_names"]
feature_names
```

```python
X = pd.DataFrame(data, columns=feature_names)
X.head()
```

```python
X.info()
```

```python
target = boston["target"]
target[:10]
```

```python
y = pd.DataFrame(target, columns=["price"])
y.info()
```

```python
#Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции 
#train_test_split так, чтобы размер тестовой выборки составлял 30% от всех данных, при этом аргумент 
#random_state должен быть равен 42.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression 
#из модуля sklearn.linear_model.
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
#Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred.shape
```

```python
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten(),
})
check_test["error"] = check_test["y_pred"] - check_test["y_test"]
check_test.head()
```

```python
#средняя квадратическая ошибка
from sklearn.metrics import mean_squared_error
mean_squared_error(check_test["y_pred"], check_test["y_test"])
```

```python
#средняя абсолютная ошибка
from sklearn.metrics import mean_absolute_error
mean_absolute_error(check_test["y_pred"], check_test["y_test"])
```

```python
#Вычислите R2 полученных предказаний с помощью r2_score из модуля sklearn.metrics.
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
#Задание 2
#Создайте модель под названием model с помощью RandomForestRegressor из модуля sklearn.ensemble.
from sklearn.ensemble import RandomForestRegressor
#Сделайте агрумент n_estimators равным 1000, max_depth должен быть равен 12 и random_state сделайте равным 42.
model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
```

```python
#Обучите модель на тренировочных данных аналогично тому, как вы обучали модель LinearRegression, но при этом 
#в метод fit вместо датафрейма y_train поставьте y_train.values[:, 0], чтобы получить из датафрейма одномерный 
#массив Numpy,так как для класса RandomForestRegressor в данном методе для аргумента y предпочтительно применение 
#массивов вместо датафрейма.
model.fit(X_train, y_train.values[:, 0])
```

```python
#Сделайте предсказание на тестовых данных и посчитайте R2. Сравните с результатом из предыдущего задания.
y_pred_rf = model.predict(X_test)
y_pred_rf.shape
```

```python
check_test_rf = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred_rf": y_pred_rf.flatten(),
})
check_test_rf["error"] = check_test_rf["y_pred_rf"] - check_test_rf["y_test"]
check_test_rf.head()
```

```python
#средняя квадратическая ошибка
from sklearn.metrics import mean_squared_error
mean_squared_error(check_test_rf["y_pred_rf"], check_test_rf["y_test"])
```

```python
#средняя абсолютная ошибка
from sklearn.metrics import mean_absolute_error
mean_absolute_error(check_test_rf["y_pred_rf"], check_test_rf["y_test"])
```

```python
#Вычислите R2 полученных предказаний с помощью r2_score из модуля sklearn.metrics.
from sklearn.metrics import r2_score
r2_score(y_test, y_pred_rf)
```

```python
#Задание 4
#Импортируйте из соответствующих модулей RandomForestClassifier, GridSearchCV и train_test_split. 
#Загрузите датасет creditcard.csv и создайте датафрейм df.
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
df = pd.read_csv("creditcard.csv")
df.head()
```

```python
#С помощью метода value_counts с аргументом normalize=True убедитесь в том, что выборка несбалансирована.
df['Class'].value_counts(normalize=True)
```

```python
#Используя метод info, проверьте, все ли столбцы содержат числовые данные и нет ли в них пропусков.
df.info()
```

```python
#Примените следующую настройку, чтобы можно было просматривать все столбцы датафрейма: 
pd.options.display.max_columns = 100
#Просмотрите первые 10 строк датафрейма df.
df.head(10)
```

```python
#Создайте датафрейм X из датафрейма df, исключив столбец Class.
X = df.drop("Class", axis=1)
#Создайте объект Series под названием y из столбца Class.
y = df["Class"]
```

```python
#Разбейте X и y на тренировочный и тестовый наборы данных при помощи функции train_test_split, используя аргументы:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
#Просмотрите информацию о их форме.
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#Для поиска по сетке параметров задайте такие параметры:
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
```

```python
clf = GridSearchCV(estimator=RandomForestClassifier(random_state=100),
param_grid=parameters,
scoring='roc_auc',
cv=3)
```

```python
#Обучите модель на тренировочном наборе данных (может занять несколько минут).
clf.fit(X_train, y_train)
```

```python
#Просмотрите параметры лучшей модели с помощью атрибута best_params_.
clf.best_params_
```

```python
#Предскажите вероятности классов с помощью полученнной модели и метода predict_proba.
y_pred_clf = clf.predict_proba(X_test)
y_pred_clf.shape
```

```python
#Из полученного результата (массив Numpy) выберите столбец с индексом 1 (вероятность класса 1) и запишите в массив y_pred_proba. 
y_pred_proba = y_pred_clf[:, 1]
print(y_pred_proba[:10])
```

```python
#Из модуля sklearn.metrics импортируйте метрику roc_auc_score.
from sklearn.metrics import roc_auc_score
#Вычислите AUC на тестовых данных и сравните с результатом,полученным на тренировочных данных, используя 
#в качестве аргументов массивы y_test и y_pred_proba.
roc_auc_score(y_test,y_pred_proba)
```
